```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Generate training X and training Y data

```python
# add TeamCleaned.csv with Abbrevs

dfall = pd.read_csv('../tmstats/TeamCleaned.csv')
dfall = dfall.drop(columns = 'Unnamed: 0')

abbrevs = pd.read_csv("../team_abbreviations.csv", 
                      header=None, 
                      names=['abbreviation','full_name'])

dfall['TeamAbbrevs']=''
for i,team in dfall.iterrows():
    for j,team_abbrev in abbrevs.iterrows():
        if '*' in team['Team']:
            a = team['Team'][:-1]
        else:
            a = team['Team']
    
        if a in team_abbrev['full_name']:
            dfall.at[i,'TeamAbbrevs'] = team_abbrev['abbreviation']
            
print('If there is null in dataset?', dfall.empty)
```

```python
tmstats = dfall

## scale the data
tmstats['W'] = tmstats['W']/tmstats['GP']
tmstats['L'] = tmstats['L']/tmstats['GP']
tmstats['GF'] = tmstats['GF']/tmstats['GP']
tmstats['GA'] = tmstats['GA']/tmstats['GP']
tmstats['S'] = tmstats['S']/tmstats['GP']
tmstats['SA'] = tmstats['SA']/tmstats['GP']
tmstats['MsS'] = tmstats['MsS']/tmstats['GP']
tmstats['TA/GA'] = tmstats['TkA/60']/tmstats['GvA/60']

print(tmstats['Year'].unique())
print(tmstats.columns)
```

```python
## combine all Positions by Game together
data = pd.read_csv('../Positions by Game/2012-2013_positions_by_game.csv')
data = data.drop(columns = 'Unnamed: 0')
column_names = data.columns

dfall = pd.DataFrame(columns = column_names)
dfall['year']=''

a = 2021
for i in range(16):
    a -= 1
    filename = '../Positions by Game/'+str(a)+'-'+str(a+1)+'_positions_by_game.csv'
    data_now = pd.read_csv(filename)
    data_now = data_now.drop(columns = 'Unnamed: 0')
    
    df = pd.DataFrame(columns = column_names)
    df['year']=''

    df[column_names]=data_now[column_names]
    df['year'] = a
    dfall = pd.concat([dfall, df], ignore_index=True)

player_data = dfall
print('If there is null in dataset?', player_data.empty)

print(player_data['year'].unique())
```

```python
## combine team and players dataset together

see_column = np.array(['Rk', 'AvAge', 'GP', 'W', 'L', 'OL', 'PTS', 'PTS%',
       'GF', 'GA', 'SOW', 'SOL', 'SRS', 'SOS', 'GF/G', 'GA/G', 'PP', 'PPO',
       'PP%', 'PPA', 'PPOA', 'PK%', 'SH', 'SHA', 'PIM/G', 'oPIM/G', 'S', 'S%',
       'SA', 'SV%', 'SO', 'Hits', 'Hits/60', 'BkS', 'BkS/60', 'GvA', 'GvA/60',
       'TkA', 'TkA/60', 'ENG', 'MsS', '5v5 TOI/GP', 'SAT%', 'Playoffs%',
       'Playoffs', 'WonCup', 'TA/GA'])

see_column2 = np.array(['homeTeam', 'awayTeam', 
                       'homeAbbrev', 'awayAbbrev', 
                       'GamePK', 'year', 'Winner',
                       'F_home_assists', 'F_home_goals', 'F_home_pim', 'F_home_shots',
                       'F_home_blocked', 'F_home_pm', 'D_home_assists', 'D_home_goals',
                       'D_home_pim', 'D_home_shots', 'D_home_blocked', 'D_home_hits',
                       'D_home_pm', 'G_home_GAA', 'F_away_assists', 'F_away_goals',
                       'F_away_pim', 'F_away_shots', 'F_away_blocked', 'F_away_hits',
                       'F_away_pm', 'D_away_assists', 'D_away_goals', 'D_away_pim',
                       'D_away_shots', 'D_away_blocked', 'D_away_hits', 'D_away_pm',
                       'G_away_GAA', 'F_home_hits'])


column_home = np.array([i+'_H' for i in see_column])
column_away = np.array([i+'_A' for i in see_column])

column_name = np.concatenate((column_home,column_away))

column_name = np.concatenate((see_column2,column_name))
#  column_name = np.concatenate((column_name,['class'])) # Winner is the class

print(column_name)
print(len(column_name))
```

```python
dfall = pd.DataFrame(columns = column_name)

for i,game in player_data.iterrows():
    year = game['year']
    home_abbrev = game['homeAbbrev']
    away_abbrev = game['awayAbbrev']
    
    tmsub_home = tmstats.loc[(tmstats['Year'] == year) &
                             (tmstats['TeamAbbrevs'] == home_abbrev)][see_column]
    
    tmsub_away = tmstats.loc[(tmstats['Year'] == year) &
                             (tmstats['TeamAbbrevs'] == away_abbrev)][see_column]
    
    if tmsub_home.empty or tmsub_away.empty:
        print(f'Eliminate this data sine data empty for team {home_abbrev} or {away_abbrev} in year {year}')
    else:
        a = game[see_column2].to_numpy()
        b = tmsub_home.to_numpy().flatten()
        c = tmsub_away.to_numpy().flatten()
        
        onerow = np.concatenate((a, b))
        onerow = np.concatenate((onerow, c))
        subdf = pd.DataFrame(np.array([onerow]), columns = column_name)
        dfall = pd.concat([dfall, subdf], ignore_index=True)
```
